```python
# ────────────────────────────
# FUNZIONE DI PLOT COMUNE

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def barplot_marketcap(df, title):
    df = df.sort_values("MarketCap_EUR_B", ascending=False).reset_index(drop=True)
    df["Label"] = df.apply(
        lambda r: f"{r['Company']} ({r['StateStake_%']:.1f}%)" if r['StateStake_%']>0 else r['Company'],
        axis=1)
    colors = ['#B89C4D' if s>0 else '#4B4E6D' for s in df["StateStake_%"]]

    fig, ax = plt.subplots(figsize=(18, 8))
    x = np.arange(len(df))
    ax.bar(x, df["MarketCap_EUR_B"], color=colors,
           edgecolor='#f0f0f0', linewidth=0.6)
    ax.margins(x=0.005)
    ax.yaxis.grid(True, color='#847d7d', linestyle='--',
                  linewidth=0.6, alpha=0.6)

    ax.set_xticks(x)
    ax.set_xticklabels(df["Label"], rotation=90, fontsize=7, ha='center')
    ax.set_ylabel("Capitalizzazione di mercato (mld €)", font='serif', fontsize=12)
    ax.set_title(title, font='serif', fontsize=15, pad=20)
    fig.text(0.01, 0.01, "Grafico prodotto da Josef di Giuseppe",
             ha='left', va='bottom', fontsize=8)
    fig.patch.set_facecolor("#f2f0f0")
    plt.tight_layout(rect=[0, 0.05, 1, 0.97])
    plt.show()



# ────────────────────────────
# 1. ITALIA – FTSE MIB 40

data_it = [

    {"Company": "UniCredit", "MarketCap_EUR_B": 96.69, "StateStake_%": 0.0},
    {"Company": "Intesa Sanpaolo", "MarketCap_EUR_B": 94.05, "StateStake_%": 0.0},
    {"Company": "Enel", "MarketCap_EUR_B": 86.34, "StateStake_%": 23.6},
    {"Company": "Ferrari", "MarketCap_EUR_B": 60.58, "StateStake_%": 0.0},
    {"Company": "Generali", "MarketCap_EUR_B": 48.11, "StateStake_%": 0.0},
    {"Company": "Eni", "MarketCap_EUR_B": 46.19, "StateStake_%": 31.835},
    {"Company": "Leonardo", "MarketCap_EUR_B": 30.18, "StateStake_%": 30.2},
    {"Company": "Stellantis", "MarketCap_EUR_B": 25, "StateStake_%": 0.0},
    {"Company": "STMicroelectronics", "MarketCap_EUR_B": 22.98, "StateStake_%": 27.5},
    {"Company": "Poste Italiane", "MarketCap_EUR_B": 25.6, "StateStake_%": 64.26},
    {"Company": "Prysmian", "MarketCap_EUR_B": 25.01, "StateStake_%": 0.0},
    {"Company": "Banco BPM", "MarketCap_EUR_B": 19.11, "StateStake_%": 0.0},
    {"Company": "Tenaris", "MarketCap_EUR_B": 17.83, "StateStake_%": 0.0},
    {"Company": "Terna", "MarketCap_EUR_B": 17.79, "StateStake_%": 29.85},
    {"Company": "Snam", "MarketCap_EUR_B": 17.30, "StateStake_%": 31.35},
    {"Company": "BPER Banca", "MarketCap_EUR_B": 18.66, "StateStake_%": 0.0},
    {"Company": "Moncler", "MarketCap_EUR_B": 14.5, "StateStake_%": 0.0},
    {"Company": "Mediobanca", "MarketCap_EUR_B": 13.31, "StateStake_%": 0.0},
    {"Company": "Banca Mediolanum", "MarketCap_EUR_B": 12.47, "StateStake_%": 0.0},
    {"Company": "Telecom Italia", "MarketCap_EUR_B": 11.0, "StateStake_%": 24.8},
    {"Company": "Recordati", "MarketCap_EUR_B": 10.66, "StateStake_%": 0.0},
    {"Company": "FinecoBank", "MarketCap_EUR_B": 10.3, "StateStake_%": 0.0},
    {"Company": "INWIT", "MarketCap_EUR_B": 9.31, "StateStake_%": 0.0},
    {"Company": "Banca Monte dei Paschi di Siena", "MarketCap_EUR_B": 22.11, "StateStake_%": 4.863},
    {"Company": "Buzzi Unicem", "MarketCap_EUR_B": 8.84, "StateStake_%": 0.0},
    {"Company": "A2A", "MarketCap_EUR_B": 7.12, "StateStake_%": 50.1},
    {"Company": "Campari", "MarketCap_EUR_B": 6.73, "StateStake_%": 0.0},
    {"Company": "Italgas", "MarketCap_EUR_B": 8.69, "StateStake_%": 25.96},
    {"Company": "Nexi", "MarketCap_EUR_B": 6.41, "StateStake_%": 0.0},
    {"Company": "Brunello Cucinelli", "MarketCap_EUR_B": 6.11, "StateStake_%": 0.0},
    {"Company": "Banca Popolare di Sondrio", "MarketCap_EUR_B": 6.04, "StateStake_%": 0.0},
    {"Company": "Pirelli", "MarketCap_EUR_B": 5.93, "StateStake_%": 0.0},
    {"Company": "Hera", "MarketCap_EUR_B": 5.67, "StateStake_%": 45.8},
    {"Company": "Saipem", "MarketCap_EUR_B": 5.48, "StateStake_%": 34.01},
    {"Company": "Iveco Group", "MarketCap_EUR_B": 4.81, "StateStake_%": 0.0},
    {"Company": "Azimut", "MarketCap_EUR_B": 4.59, "StateStake_%": 0.0},
    {"Company": "Interpump Group", "MarketCap_EUR_B": 4.10, "StateStake_%": 0.0},
    {"Company": "DiaSorin", "MarketCap_EUR_B": 4.06, "StateStake_%": 0.0},
    {"Company": "Amplifon", "MarketCap_EUR_B": 3.35, "StateStake_%": 0.0},
    {"Company": "Unipol", "MarketCap_EUR_B": 12.92, "StateStake_%": 0.0}

]

df_it = pd.DataFrame(data_it, columns=["Company", "MarketCap_EUR_B", "StateStake_%"])
barplot_marketcap(df_it, "FTSE MIB 40 2025 – Market cap e quota pubblica (MEF, CDP, altri)")

# ────────────────────────────
# 2. FRANCIA – CAC 40


data_fr = [
    ("LVMH", 351, 0),
    ("Hermès", 269, 0),
    ("L'Oréal", 243, 0),
    ("TotalEnergies", 132, 0),
    ("Sanofi", 122, 0),
    ("Airbus", 184, 11),
    ("Schneider Electric", 161, 0),
    ("BNP Paribas", 97, 0),
    ("EssilorLuxottica", 167, 0),
    ("Air Liquide", 116, 0),
    ("Kering", 45, 0),
    ("Safran", 145, 11),
    ("Vinci", 79, 0),
    ("Dassault Systèmes", 45, 0),
    ("Capgemini", 23, 0),
    ("Engie", 55, 23.6),
    ("Bouygues", 18, 0),
    ("ArcelorMittal", 29, 0),
    ("Orange", 44, 23),
    ("STMicroelectronics", 26, 13.8),
    ("Veolia", 25, 0),
    ("Thales", 59, 25.7),
    ("Pernod Ricard", 24, 0),
    ("Publicis", 25, 0),
    ("Danone", 58, 0),
    ("Renault", 11, 15),
    ("Euronext", 15, 0),
    ("Carrefour", 11, 0),
    ("Saint‑Gobain", 51, 0),
    ("Legrand", 44, 0),
    ("Crédit Agricole", 58, 0),
    ("Accor", 10, 0),
    ("Edenred", 5, 0),
    ("Eurofins Scientific", 13, 0),
    ("Bureau Veritas", 12, 0),
    ("Société Générale", 13, 0),
    ("Michelin", 22, 0),
    ("URW (Unibail‑Rodamco‑Westfield)", 15, 0),
    ("Stellantis", 25, 0),
    ("AXA", 98, 0),
]

df_fr = pd.DataFrame(data_fr, columns=["Company", "MarketCap_EUR_B", "StateStake_%"])
barplot_marketcap(df_fr, "CAC 40 2025 – Market cap e quota pubblica (APE, Bpifrance, altri)")

# ─────────────────────────────
# 3 · GERMANIA – DAX 40 (set 2025)


data_de = [
    # Company,  Market-cap € bn,  Public stake %
    ("SAP", 318, 0),
    ("Siemens", 217, 0),
    ("Allianz", 156, 0),
    ("Airbus", 184, 10.9),           
    ("Volkswagen", 53, 11.0),
    ("Mercedes-Benz Group", 59, 0),
    ("Deutsche Telekom", 170, 28.3),  
    ("BMW", 57, 0),
    ("Infineon", 50, 0),
    ("Siemens Healthineers", 62, 0),
    ("BASF", 45, 0),
    ("Bayer", 30, 0),
    ("Adidas", 39, 0),
    ("RWE", 34, 0),
    ("Deutsche Post DHL", 52, 17.0), 
    ("Deutsche Börse", 47, 0),
    ("Siemens Energy", 94, 0),
    ("Heidelberg Cements", 39, 0),
    ("Münchener Rück", 82, 0),
    ("Porsche AG", 45, 0),
    ("Henkel", 33, 0),
    ("Porsche SE", 12, 0),
    ("Fresenius Medical Care", 17, 0),
    ("Hannover Rück", 35, 0),
    ("MTU Aero Engines", 22, 0),
    ("Symrise", 14, 0),
    ("Brenntag", 8, 0),
    ("Continental", 14, 0),
    ("Daimler Truck", 30, 0),
    ("Covestro", 13, 0), 
    ("Sartorius", 18, 0),
    ("Beiersdorf", 24, 0),
    ("Deutsche Bank", 64, 0),
    ("Commerzbank", 39, 12),
    ("Qiagen", 10, 0),  
    ("Merck", 56, 0), 
    ("Rheinmetall", 90, 0),
    ("E.ON", 49, 0),
    ("Vonovia", 26, 0),
    ("Zalando", 7, 0)
]

df_de = pd.DataFrame(data_de, columns=["Company", "MarketCap_EUR_B", "StateStake_%"])
barplot_marketcap(df_de, "DAX 40 2025 – Market cap e quota pubblica (Bund, KfW, altri)")

```

```python
def stats(df):
    cap_total = df["MarketCap_EUR_B"].sum()
    cap_public = (df["MarketCap_EUR_B"] * df["StateStake_%"]/100).sum()
    pct = cap_public / cap_total * 100
    return cap_total, cap_public, pct

stats_it = stats(df_it)
stats_fr = stats(df_fr)
stats_de = stats(df_de)

summary = pd.DataFrame({
    "Indice": ["FTSE MIB 40","CAC 40","DAX 40"],
    "Cap totale (mld €)": [round(stats_it[0],1), round(stats_fr[0],1), round(stats_de[0],1)],
    "Cap controllata (mld €)": [round(stats_it[1],1), round(stats_fr[1],1), round(stats_de[1],1)],
    "% su totale": [round(stats_it[2],1), round(stats_fr[2],1), round(stats_de[2],1)]
})

summary

```

```python
fig, ax = plt.subplots(figsize=(9, 2))
ax.axis('off')

#Tabella Grafica
table = ax.table(cellText=summary.values,
                 colLabels=summary.columns,
                 cellLoc='center',
                 loc='center')

table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1, 2)  #Aggiusta dimensione tabella

plt.show()
```
